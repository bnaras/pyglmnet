```python
import numpy as np
rng = np.random.default_rng(0)
n, p = 200, 4
X = rng.standard_normal((n, p))
Y = rng.choice(['A','B', 'C'], size=n)
```

```python
from glmnet import MultiClassNet
nlambda = 100
L = MultiClassNet(nlambda=nlambda)
L.fit(X, Y)
```

```python
import rpy2
%load_ext rpy2.ipython
%R -i X,Y,nlambda
```

```python
%%R
library(glmnet)
y = as.factor(Y)
X = as.matrix(X)
G = glmnet(X, y, family='multinomial', nlam=nlambda)
#plot(G)
G
```

```python
%%R -o C_R1,C_R2,C_R3,I_R1,I_R2,I_R3
C_R1 = as.matrix(coef(G)$A)[-1,]
C_R2 = as.matrix(coef(G)$B)[-1,]
C_R3 = as.matrix(coef(G)$C)[-1,]
I_R1 = as.matrix(coef(G)$A)[1,]
I_R2 = as.matrix(coef(G)$B)[1,]
I_R3 = as.matrix(coef(G)$C)[1,]
```

```python
C_R = np.concatenate([C_R1.T[:,:,None], C_R2.T[:,:,None], C_R3.T[:,:,None]], axis=2)
C_R.shape
```

```python
np.linalg.norm(C_R - L.coefs_) / np.linalg.norm(L.coefs_)
```

```python
I_R = np.array([I_R1,
                I_R2,
                I_R3])
np.linalg.norm(I_R.T - L.intercepts_) / np.linalg.norm(L.intercepts_)
```
